```python
# consider only important columns:
# keep only the following columns in your dataframe- Date, District, Count(our label), Year
#### insert your code here:
theft_features = theft_data[['Date', 'District', 'Count', 'Year']]
theft_features
# Expected Result: Index(['Date', 'District', 'Count', 'Year'], dtype='object')
```

### Feature extraction:

```python
# change the Date from string to timestamp, in theft_data & holidays_data_bgu:
theft_features['Date'] = pd.to_datetime(theft_features['Date'], format = '%Y-%m-%d')
theft_features.Date
```

```python
# add time information create day of week column:
# name the column Week_day, it will contain a number for each day of week:
# Monday=0, ...., Sunday=6

from datetime import date
theft_features.loc[:,'DayOfWeek'] = theft_features['Date'].dt.weekday
theft_features
```

```python
from datetime import date
# and write function that gets the date and returns the corresponding season
# Spring start 21.3
# Summer start 21.6
# Autumn start 23.9
# Winter start 21.12

def Season(date):
  month = date.month
  day = date.day
  if (month == 12 and day >= 21) or (month in [1, 2]) or (month == 3 and day < 20):
        return 'Winter'
  elif (month == 3 and day >= 20) or (month in [4, 5]) or (month == 6 and day < 21):
        return 'Spring'
  elif (month == 6 and day >= 21) or (month in [7, 8]) or (month == 9 and day < 23):
        return 'Summer'
  elif (month == 9 and day >= 23) or (month in [10, 11]) or (month == 12 and day < 21):
        return 'Autumn'
  else:
        return None

# name the column Season, it will contain a string which tells us the season based on the Date:
theft_features.loc[:,'Season'] = theft_features.loc[:,'Date'].apply(Season)
theft_features

```

```python
# add information from other sources: holidays data
# create column is_holiday that will take 1 if its an holiday and 0 otherwise.
# hint: you can use "merge" operation
import pandas as pd
holidays_data_bgu['Date'] = pd.to_datetime(holidays_data_bgu['Date'], format = 'mixed')
merged_theft_features = pd.merge(theft_features, holidays_data_bgu[['Date', 'is_holiday']], on='Date', how='left').fillna(0)
merged_theft_features['is_holiday'] = merged_theft_features.is_holiday.astype(int)
merged_theft_features
```

```python
# now you suppose to have the following columns: Date, is_holiday, district, count(our label), year, week_day, season
# expected result: 'Date', 'is_holiday', 'District', 'Count', 'Year', 'Week_day','Season'
merged_theft_features.columns
merged_theft_features
```

```python
# create dummy variables:
# hint - use the pandas function 'get_dummies' with drop_first=True
#### insert your code here:
pd.get_dummies(merged_theft_features, columns=['Season'], drop_first=True)
####
```

```python
# now you suppose to have the following columns: Date, is_holiday, count, year,
# dummies for season, week_day and district
pd.get_dummies(merged_theft_features, columns=['Season','DayOfWeek','District'], drop_first=True)
#merged_theft_features.columns
```

```python
# remove the original date columns: (we won't use it for modelling)
total_theft_data = pd.get_dummies(merged_theft_features, columns=['Season','DayOfWeek','District'], drop_first=True).drop(['Date'],axis=1)
total_theft_data
total_theft_data['Year']==2016
```

### Train-Test Splitting:

```python
# choose years for train and test
train_start_year = 2014
train_end_year = 2015
test_year = 2016
```

```python
# split the data into train/test

# choosing the data with rows that match ONLY years 2014-2015 for training data
dataTrain = total_theft_data[(total_theft_data["Year"] >= train_start_year) & (total_theft_data["Year"] <= train_end_year)]
# keeping count column from data frame
labelsTrain = dataTrain.Count
# droping count column from the training data
dataTrain = dataTrain.drop('Count', axis=1)

# choosing the data with rows that match ONLY year 2016 for testing data
dataTest = total_theft_data[(total_theft_data["Year"] == test_year)]
# keeping count column from data frame
labelsTest = dataTest.Count
# droping count column from the training data
dataTest = dataTest.drop('Count', axis=1)

# Remove unnecessary columns from training data and testing data:
dataTrain = dataTrain.drop('Year', axis=1)
dataTest = dataTest.drop('Year', axis=1)


print("Train data shape: " , dataTrain.shape)
print("Test data shape: " , dataTest.shape)
```

```python
# check for null values (should print 0)
# for 0 there are no null values in our data
print(total_theft_data.isnull().sum().sum())
```

### Modelling

```python
from sklearn.linear_model import LinearRegression
mlModel = LinearRegression()
mlModel.fit(dataTrain, labelsTrain)
```

### Evaluation

```python
predTest = mlModel.predict(dataTest)

# print the Rsquare:
print("Test set R^2: ", mlModel.score(dataTest, labelsTest))
```
